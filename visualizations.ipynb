```python
def match_particle(hit):
    particle_id = truth.loc[truth["hit_id"]==hit]["particle_id"].iloc[0]
    mass = truth.loc[truth["hit_id"]==hit]["weight"].iloc[0]
    particle = particles.loc[particles["particle_id"] == particle_id]
    return (particle, mass)
    truth.loc[ truth["columnanme"] == "some values"]
match_particle(6)
```

```python
def unpacker(value):
    #This step is crucial
    list_val = list(value)
    if len(list_val) < 1:
        return 0
    else:
        return list_val[0]
```

```python
# goose = pd.DataFrame([["one", "two"], ["two", "Three"]], columns=["one", "Two"])
# display(goose)

# duck = pd.DataFrame([["duck", "tuck"], ["muck", "luck"]], columns=["ruckus", "muckus"])
# display(duck)

# birds = [goose, duck]

# fritters = pd.concat(birds)
# display(fritters)

###
#section_parts_df(df, x_min = False, x_max=False, y_min=False, x_max=False, z_min=False, z_max=False ):
#    for index in range(0, len(df)):
        
###
print(len(hits))
def build_display_df(num_particles = None):
    if num_particles == None:
        num_particles = len(hits)

    p_cols = ["x", "y", "z", "q", "m", "v"]
    p_list = [] 

    max_m = 0
    min_m = 999999999
    
    for row in range(0, num_particles):
        particle_tuple = match_particle(hits.iloc[row]["hit_id"])
        particle = particle_tuple[0]
        m = particle_tuple[1]*1000000

        q = unpacker(particle["q"])

#         vx = unpacker(particle["vx"])
#         vy = unpacker(particle["vy"])
#         vz = unpacker(particle["vz"])
#         v = math.sqrt(math.pow(vx,2)+math.pow(vy,2)+math.pow(vz,2))

        px = unpacker(particle["px"])
        p_y = unpacker(particle["py"])
        pz = unpacker(particle["pz"])
        p = math.sqrt(math.pow(px,2)+math.pow(p_y,2)+math.pow(pz,2))

        # p = mv
        # tanh scale for visualization
        
#         if v > 0:
#             m = p/v
#         else:
#             m = 0

#         if m > max_m:
#             max_m = m
#         if m < min_m:
#             min_m = m
        row = [hits.iloc[row]["x"], hits.iloc[row]["y"], hits.iloc[row]["z"], q, m, p]
        p_list.append(row)
        
    return pd.DataFrame(p_list, columns=p_cols)
```

```python
def plot_hist(column, labels):
    print("{} unique values".format(len(column)))
    fig = ff.create_distplot(column, labels)
    display(py.iplot(fig, filename='histogram'+str(labels)))
```

```python
def charge_plot(x,y,z,q):
    trace = go.Scatter3d(
    x=y, # for orientation with pyplot visualization
    y=z,
    z=x,
    mode='markers',
    marker=dict(
            size=2,
            color=q,   # set color 
            colorscale='Viridis',   #colorscale
            opacity=0.2
        )
    )
    layout = go.Layout(showlegend=True)
    fig = go.Figure(data=[trace], layout=layout)
    py.iplot(fig, filename='hits'+q)
```

```python
def scatter3d_color(x, y, z, c, t):
    
    print("{} unique values".format({len(x), len(y), len(z)}))
    matrace = go.Scatter3d(
            x=x,
            y=y,
            z=z,
            mode='markers',
            marker=dict(
                    size=2,
                    color=c,   # set color to an array/list of desired values
                    colorscale='Viridis',   # choose a colorscale
                    opacity=0.2
        )
    )
    layout = go.Layout(showlegend=True)
    fig = go.Figure(data=[matrace], layout=layout)
    display(type(py))
    display(py.iplot(fig, filename='hits'+t))
```
